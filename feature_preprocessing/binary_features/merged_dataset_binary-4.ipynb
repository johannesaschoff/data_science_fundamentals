```python
import pandas as pd
import numpy as np

timeframe = pd.read_csv("data/accurate_timeframe.csv")
timeframe.set_index("datetime", inplace = True)
timeframe.rename_axis("Date", inplace = True)

# Hourly Temperature, Weather & Sunshine

weather = pd.read_csv("data/weather.csv")
weather.set_index("Date", inplace = True)

# Public & School Holidays of Neighbor Cantons

## ZH

zh_public = pd.read_csv("data/zh_public.csv")
zh_public.set_index("Date", inplace = True)

zh_school = pd.read_csv("data/zh_school_holidays.csv")
zh_school.set_index("Date", inplace = True)

## SH

sh_public = pd.read_csv("data/sh_public.csv")
sh_public.set_index("Date", inplace = True)

sh_school = pd.read_csv("data/sh_school_holidays.csv")
sh_school.set_index("Date", inplace = True)

## TG

tg_public = pd.read_csv("data/tg_public.csv")
tg_public.set_index("Date", inplace = True)

tg_school = pd.read_csv("data/tg_school_holidays.csv")
tg_school.set_index("Date", inplace = True)

## SG

sg_public = pd.read_csv("data/sg_public.csv")
sg_public.set_index("Date", inplace = True)

sg_school = pd.read_csv("data/sg_school_holidays.csv")
sg_school.set_index("Date", inplace = True)

## SG

sz_public = pd.read_csv("data/sz_public.csv")
sz_public.set_index("Date", inplace = True)

sz_school = pd.read_csv("data/sz_school_holidays.csv")
sz_school.set_index("Date", inplace = True)

## ZG

zg_public = pd.read_csv("data/zg_public.csv")
zg_public.set_index("Date", inplace = True)

zg_school = pd.read_csv("data/zg_school_holidays.csv")
zg_school.set_index("Date", inplace = True)

## AG

ag_public = pd.read_csv("data/ag_public.csv")
ag_public.set_index("Date", inplace = True)

ag_school = pd.read_csv("data/ag_school_holidays.csv")
ag_school.set_index("Date", inplace = True)

## Stores Closed

stores_closed = pd.read_csv("data/stores_closed.csv")
stores_closed["datetime"] = pd.to_datetime(stores_closed["datetime"])
stores_closed.drop_duplicates(inplace=True)
stores_closed.set_index("datetime", inplace = True)
stores_closed.rename_axis("Date", inplace = True)

## Influential Events

influeantial_events = pd.read_csv("data/infulential_events.csv")
influeantial_events["datetime"] = pd.to_datetime(influeantial_events["datetime"])
influeantial_events.set_index("datetime", inplace = True)
influeantial_events.rename_axis("Date", inplace = True)
```

```python

## ZH

merged_features = pd.merge(timeframe, zh_public[["Public Holiday ZH"]], left_index=True, right_index=True, how="left")
merged_features["Public Holiday ZH"] = merged_features["Public Holiday ZH"].fillna(False).astype(bool).astype(float)

merged_features = pd.merge(merged_features, zh_school[["School Holidays ZH"]], left_index=True, right_index=True, how="left")
merged_features["School Holidays ZH"] = merged_features["School Holidays ZH"].fillna(False).astype(bool).astype(float)

## SH

merged_features = pd.merge(merged_features, sh_public[["Public Holiday SH"]], left_index=True, right_index=True, how="left")
merged_features["Public Holiday SH"] = merged_features["Public Holiday SH"].fillna(False).astype(bool).astype(float)

merged_features = pd.merge(merged_features, sh_school[["School Holidays SH"]], left_index=True, right_index=True, how="left")
merged_features["School Holidays SH"] = merged_features["School Holidays SH"].fillna(False).astype(bool).astype(float)

## TG

merged_features = pd.merge(merged_features, tg_public[["Public Holiday TG"]], left_index=True, right_index=True, how="left")
merged_features["Public Holiday TG"] = merged_features["Public Holiday TG"].fillna(False).astype(bool).astype(float)

merged_features = pd.merge(merged_features, tg_school[["School Holidays TG"]], left_index=True, right_index=True, how="left")
merged_features["School Holidays TG"] = merged_features["School Holidays TG"].fillna(False).astype(bool).astype(float)

## SG

merged_features = pd.merge(merged_features, sg_public[["Public Holiday SG"]], left_index=True, right_index=True, how="left")
merged_features["Public Holiday SG"] = merged_features["Public Holiday SG"].fillna(False).astype(bool).astype(float)

merged_features = pd.merge(merged_features, sg_school[["School Holidays SG"]], left_index=True, right_index=True, how="left")
merged_features["School Holidays SG"] = merged_features["School Holidays SG"].fillna(False).astype(bool).astype(float)

## SZ

merged_features = pd.merge(merged_features, sz_public[["Public Holiday SZ"]], left_index=True, right_index=True, how="left")
merged_features["Public Holiday SZ"] = merged_features["Public Holiday SZ"].fillna(False).astype(bool).astype(float)

merged_features = pd.merge(merged_features, sz_school[["School Holidays SZ"]], left_index=True, right_index=True, how="left")
merged_features["School Holidays SZ"] = merged_features["School Holidays SZ"].fillna(False).astype(bool).astype(float)

## ZG

merged_features = pd.merge(merged_features, zg_public[["Public Holiday ZG"]], left_index=True, right_index=True, how="left")
merged_features["Public Holiday ZG"] = merged_features["Public Holiday ZG"].fillna(False).astype(bool).astype(float)

merged_features = pd.merge(merged_features, zg_school[["School Holidays ZG"]], left_index=True, right_index=True, how="left")
merged_features["School Holidays ZG"] = merged_features["School Holidays ZG"].fillna(False).astype(bool).astype(float)

## AG

merged_features = pd.merge(merged_features, ag_public[["Public Holiday AG"]], left_index=True, right_index=True, how="left")
merged_features["Public Holiday AG"] = merged_features["Public Holiday AG"].fillna(False).astype(bool).astype(float)

merged_features = pd.merge(merged_features, ag_school[["School Holidays AG"]], left_index=True, right_index=True, how="left")
merged_features["School Holidays AG"] = merged_features["School Holidays AG"].fillna(False).astype(bool).astype(float)

# Hourly Temperature, Weather & Sunshine

merged_features = pd.merge(merged_features, weather, left_index=True, right_index=True, how="left")

# Weekday

merged_features.reset_index(inplace = True)
merged_features["Date"] = pd.to_datetime(merged_features["Date"])
merged_features["Weekday"] = merged_features["Date"].dt.strftime("%A")
merged_features.set_index("Date", inplace = True)
merged_features = pd.get_dummies(merged_features, columns=["Weekday"], prefix = "Weekday")

# Stores Closed

merged_features = pd.merge(merged_features, stores_closed[["stores_closed_on_that_day"]], left_index=True, right_index=True, how="left")

## Influential Events

merged_features = pd.merge(merged_features, influeantial_events[["Event"]], left_index=True, right_index=True, how="left")
merged_features.reset_index(inplace = True)

merged_features

```

```python
# Reordering the DataFrame

order = ["Date", 
         "stores_closed_on_that_day",
         "Event",
         "clear-night",
         "cloudy",
         "fog",
         "partly-cloudy-day",
         "partly-cloudy-night",
         "rain",
         "snow",
         "wind",
         "Public Holiday ZH", 
         "Public Holiday SH", 
         "Public Holiday TG", 
         "Public Holiday SG", 
         "Public Holiday SZ", 
         "Public Holiday ZG", 
         "Public Holiday AG", 
         "School Holidays ZH", 
         "School Holidays SH", 
         "School Holidays TG", 
         "School Holidays SG", 
         "School Holidays SZ", 
         "School Holidays ZG", 
         "School Holidays AG", 
         "Weekday_Monday", 
         "Weekday_Tuesday", 
         "Weekday_Wednesday", 
         "Weekday_Thursday", 
         "Weekday_Friday", 
         "Weekday_Saturday", 
         "Weekday_Sunday"]

merged_features = merged_features[order]

# Setting the Date Range for the Testing

merged_features["stores_closed_on_that_day"] = merged_features["stores_closed_on_that_day"].fillna(0)

merged_features.set_index("Date", inplace = True)

merged_features.to_csv("binary_features.csv")
merged_features
```
